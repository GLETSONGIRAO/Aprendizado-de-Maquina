# Algoritmo Gradiente Descendente

```python
df = pd.read_csv('../data/artificial1d.csv', names = ['x','y'])
X = df['x'].to_numpy().reshape(-1,1)
y = df['y'].to_numpy()
```

```python
alfa = 0.01
X = np.c_[np.ones(X.shape[0]), X]
w = np.array([0.33, 0.55])
#w = np.random.rand(X.shape[1])
t = 1000
```

```python
w_passados = []
custos = []
for epoca in range(t):
    #X = 30,2 , w = (2,) X @ w = (30,)
    pred = X @ w
    erro = y - pred
    # Calculando o custo
    custo = (1/(2*len(y))) * np.sum((y - pred)**2)
    # Calculando os parâmetros
    # w = (2,0) , X.T = (2,30) , erro = (30,) , X.T @ erro = (2,0)
    w = w + (alfa * (1/len(y)) * (X.T @ erro))
    
    custos.append(custo)
    w_passados.append(w)
    print(w)
    
    
    
    
```

```python
custos
```

```python
class RegressaoGD():
    def __init__(self, t=1000, taxa = 0.01):
        
        self.w = np.array([0.33, 0.55])
        self.t = t
        self.taxa = taxa
        self.w_passados = []
        self.custos = []
        
    def fit(self, X, y):
        X = np.c_[np.ones(X.shape[0]), X]
        
        for epoca in range(self.t):
            #X = 30,2 , w = (2,) X @ w = (30,)
            pred = X @ self.w
            erro = y - pred
            # Calculando o custo
            custo = (1/(2*len(y))) * np.sum((y - pred)**2)
            # Calculando os parâmetros
            # w = (2,0) , X.T = (2,30) , erro = (30,) , X.T @ erro = (2,0)
            self.w = self.w + (self.taxa * (1/len(y)) * (X.T @ erro))
            #Preenchendo as listas com o histórico
            self.custos.append(custo)
            self.w_passados.append(self.w)
            
    def predict(self, X):
        X = np.c_[np.ones(X.shape[0]), X]
        
        return X @ self.w
        
    
```

```python
model = RegressaoGD()
```

```python
model.fit(X,y)
```

```python
y_pred = model.predict(X)
```
